```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sympy as sp
from google.colab import files

# Define the grid range
x_min, x_max, y_min, y_max = 0, 10, 0, 10
n_points = 100

# Create the grid of points
x = np.linspace(x_min, x_max, n_points)
y = np.linspace(y_min, y_max, n_points)
x, y = np.meshgrid(x, y)

# Function to plot and save surfaces
def plot_and_save_surfaces(x, y, z_top, z_bottom, top_file, bottom_file):
    # Plot the 3D models
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z_top, cmap='viridis', alpha=0.6)
    ax.plot_surface(x, y, z_bottom, cmap='plasma', alpha=0.6)

    # Add axis labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()

    # Convert data to XYZ format for both surfaces
    xyz_data_top = np.column_stack((x.ravel(), y.ravel(), z_top.ravel()))
    xyz_data_bottom = np.column_stack((x.ravel(), y.ravel(), z_bottom.ravel()))

    # Save the data to text files
    df_top = pd.DataFrame(xyz_data_top, columns=['X', 'Y', 'Z'])
    df_bottom = pd.DataFrame(xyz_data_bottom, columns=['X', 'Y', 'Z'])

    df_top.to_csv(top_file, index=False, header=True, sep=' ')
    df_bottom.to_csv(bottom_file, index=False, header=True, sep=' ')

    # Download the files using Google Colab
    files.download(top_file)
    files.download(bottom_file)

# Group 1
z_top_1 = np.sin(x) * np.cos(y) + 5
z_bottom_1 = np.cos(x) * np.sin(y)
plot_and_save_surfaces(x, y, z_top_1, z_bottom_1, 'top_surface1.xyz', 'bottom_surface1.xyz')

# Analytical volume calculation for Group 1
x_sym, y_sym = sp.symbols('x y')
z_top_sym_1 = sp.sin(x_sym) * sp.cos(y_sym) + 5
z_bottom_sym_1 = sp.cos(x_sym) * sp.sin(y_sym)
integrand_1 = z_top_sym_1 - z_bottom_sym_1
volume_analytic_1 = sp.integrate(integrand_1, (x_sym, x_min, x_max), (y_sym, y_min, y_max))
volume_analytic_value_1 = volume_analytic_1.evalf()
print(f'The analytical volume between the surfaces (Group 1) is: {volume_analytic_value_1}')

# Group 2
z_top_2 = x**2 + y**2 + np.sin(x + y) + 5
z_bottom_2 = x**2 - y**2 + np.cos(x - y)
plot_and_save_surfaces(x, y, z_top_2, z_bottom_2, 'top_surface2.xyz', 'bottom_surface2.xyz')

# Analytical volume calculation for Group 2
z_top_sym_2 = x_sym**2 + y_sym**2 + sp.sin(x_sym + y_sym) + 5
z_bottom_sym_2 = x_sym**2 - y_sym**2 + sp.cos(x_sym - y_sym)
integrand_2 = z_top_sym_2 - z_bottom_sym_2
volume_analytic_2 = sp.integrate(integrand_2, (x_sym, x_min, x_max), (y_sym, y_min, y_max))
volume_analytic_value_2 = volume_analytic_2.evalf()
print(f'The analytical volume between the surfaces (Group 2) is: {volume_analytic_value_2}')

# Group 3
z_top_3 = np.exp(0.1 * x) * np.sin(y) + 10
z_bottom_3 = np.exp(-0.1 * y) * np.cos(x)
plot_and_save_surfaces(x, y, z_top_3, z_bottom_3, 'top_surface3.xyz', 'bottom_surface3.xyz')

# Analytical volume calculation for Group 3
z_top_sym_3 = sp.exp(0.1 * x_sym) * sp.sin(y_sym) + 10
z_bottom_sym_3 = sp.exp(-0.1 * y_sym) * sp.cos(x_sym)
integrand_3 = z_top_sym_3 - z_bottom_sym_3
volume_analytic_3 = sp.integrate(integrand_3, (x_sym, x_min, x_max), (y_sym, y_min, y_max))
volume_analytic_value_3 = volume_analytic_3.evalf()
print(f'The analytical volume between the surfaces (Group 3) is: {volume_analytic_value_3}')
```
